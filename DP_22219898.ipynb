````python
class Person:
    def __init__(self, firstname, secondname, postcode, age):
        self._firstname = firstname
        self._secondname = secondname
        self._postcode = postcode
        self._age = age

    @property
    def firstname(self):
        return self._firstname
    
    @property
    def secondname(self):
        return self._secondname

    @property
    def postcode(self):
        return self._postcode

    @property
    def age(self):
        return self._age
       
    def friendly_message(self):
        return "Hello, my name is " + self.firstname + " "  + self.secondname + ". I am " + str(self.age) + " years old and my postcode is " + self.postcode

class Student(Person):
    def __init__(self, firstname, lastname, postcode, age, degree_subject, student_ID):
        super().__init__(firstname, lastname, postcode, age)
        self._degree_subject = degree_subject
        self._student_ID = student_ID

    @property
    def degree_subject(self):
        return self._degree_subject

    @property
    def student_ID(self):
        return self._student_ID

    def studentGreeting(self):
        return "My student ID is " + self.student_ID + " and I am reading " + self.degree_subject


p1 = Person("Freddy", "Jones", "HP6 7AJ", 22)
print(p1.friendly_message())

s1 = Student("Freddy", "Jones", "HP6 7AJ", 22, "Computer Science", "SN123456")
print(s1.studentGreeting())

print()

students = [Student("Dick", "Turpin", "HP11 2JZ", 32, "Highway Robbery", "DT123456"), Student("Dorothy", "Turpin", "SO14 7AA", 32, "Law", "DT123457"), Student("Oliver", "Cromwell", "Cromwell", 32, "History", "OC123456")]

x=0

for i in students:
    s2 = students[x]
    print(s2.friendly_message())
    print(s2.studentGreeting())
    x=x+1
````

````python
#MVC_Template_01
#2014 May 23  by Steven Lipton http://makeAppPie.com
#Controller initializing MVC -- simplest version possible.
#Additional annotations by R Mather - Oct 2018

# MVC EXTENDED WITH DELETE BUTTON

# Tkinter is a Python binding to the Tk GUI toolkit/library - RM
from tkinter import *
 
# A Model-View-Controller framework for TKinter.
# Model: Data Structure. Controller can send messages to it, and model can respond to message.
# View : User interface elements. Controller can send messages to it. View can call methods from Controller when an event happens.
# Controller: Ties View and Model together. turns UI responses into chages in data.
 


#Controller: Ties View and Model together.
#       --Performs actions based on View events.
#       --Sends messages to Model and View and gets responses
#       --Has Delegates - RM NOTE: essential for communications between view and model via control
 
class MyController():
    def __init__(self,parent):
        self.parent = parent
        # CRITICAL PART OF THE MVC 
        # RM NOTES:
        # MyController passes a reference to itself (self) when it creates MyModel and MyView objects
        # This allows the MyModel and MyView objects to create 'delegates' (see 'vc' = virtual controller?)
        # The vc objects can then call methods on the MyControl object, thereby allowing the View to update the Model via the Controller 
        self.model = MyModel(self)    # initializes the model
        self.view = MyView(self)  #initializes the view
        # Initialize objects in view
        self.view.setEntry_text('Add to Label') # a non cheat way to do MVC with tkinter control variables
        self.view.setLabel_text('Ready')
    
    #event handlers
    def quitButtonPressed(self):
        self.parent.destroy()
    def addButtonPressed(self):
        self.view.setLabel_text(self.view.entry_text.get() + ' added')
        # CRITICAL PART OF THE MVC 
        # RM NOTES - In one operation:
        #[1] Get the text from the VIEW;
        #[2] Add it to the MODEL by appending it to the list
        self.model.addToList(self.view.entry_text.get())
    #added function for when the remove button is pressed
    def removeButtonPressed(self):
        self.view.setLabel_text(self.view.entry_text.get() + ' removed')
        self.model.removeFromList(self.view.entry_text.get())
    
    def listChangedDelegate(self):
        #model internally changes and needs to signal a change
        print(self.model.getList())
 
#View : User interface elements.
#       --Controller can send messages to it.
#       --View can call methods from Controller vc when an event happens.
#       --NEVER communicates with Model.
#       --Has setters and getters to communicate with controller
 
class MyView(Frame):
    
    def __init__(self,vc):
        self.frame=Frame()
        self.frame.grid(row = 0,column=0)
        self.vc = vc
        self.entry_text = StringVar()
        self.entry_text.set('nil')
        self.label_text = StringVar()
        self.label_text.set('nil')
        self.loadView()

    def loadView(self):
        quitButton = Button(self.frame,text = 'Quit', command= self.vc.quitButtonPressed).grid(row = 0,column = 0)
        addButton = Button(self.frame,text = "Add", command = self.vc.addButtonPressed).grid(row = 0, column = 1)
        #added remove button to GUI
        removeButton = Button(self.frame,text = "Remove", command = self.vc.removeButtonPressed).grid(row = 0, column = 2)
        entry = Entry(self.frame,textvariable = self.entry_text).grid(row = 1, column = 0, columnspan = 3, sticky = EW)
        label = Label(self.frame,textvariable = self.label_text).grid(row = 2, column = 0, columnspan = 3, sticky = EW)

    def getEntry_text(self):
    #returns a string of the entry text
        return self.entry_text.get()
    def setEntry_text(self,text):
    #sets the entry text given a string
        self.entry_text.set(text)
    def getLabel_text(self):
    #returns a string of the Label text
        return self.label_text.get()
    def setLabel_text(self,text):
    #sets the label text given a string
        self.label_text.set(text)
 
#Model: Data Structure.
#   --Controller can send messages to it, and model can respond to message.
#   --Uses delegates from vc to send messages to the Control of internal change
#   --NEVER communicates with View
#   --Has setters and getters to communicate with Controller
 
class MyModel():
    def __init__(self,vc):
        self.vc = vc
        self.myList = ['duck','duck','goose']
        self.count = 0
#Delegates-- Model would call this on internal change
    def listChanged(self):
        self.vc.listChangedDelegate()
#setters and getters
    def getList(self):
        return self.myList
    def initListWithList(self, aList):
        self.myList = aList
    def addToList(self,item):
        print("returned")
        myList = self.myList
        myList.append(item)
        self.myList=myList
        self.listChanged()
    #remove button edits the list to remove item
    def removeFromList(self,item):
        print("returned")
        myList = self.myList
        myList.remove(item)
        self.myList=myList
        self.listChanged()
 
def main():
    # Create a root window from the Tk GUI class
    root = Tk()
    frame = Frame(root,bg='#0555ff' )
    root.title('Hello Penguins')
    app = MyController(root)
    root.mainloop()
 
if __name__ == '__main__':
    main()
````

````python
import matplotlib.pyplot as plt
import numpy as np


# Nicely abstracted structure by which any model can notify any observer (view) of changes in the model
class Subject(object): #Represents what is being 'observed'

    def __init__(self):
        self._observers = [] # This where references to all the observers are being kept
                             # Note that this is a one-to-many relationship: there will be one subject to be observed by multiple _observers

    def attach(self, observer):
        if observer not in self._observers: #If the observer is not already in the observers list
            self._observers.append(observer) # append the observer to the list

    def detach(self, observer): #Simply remove the observer
        try:
            self._observers.remove(observer)
        except ValueError:
            pass

    def notify(self, modifier=None):
        for observer in self._observers: # For all the observers in the list
            if modifier != observer: # Don't notify the observer who is actually doing the updating 
                observer.update(self) # Alert the observers!
                
# Represents the 'data' for which changes will produce notifications to any registered view/observer objects 
class Model(Subject): # Extends the Subject class

    def __init__(self, name):
        Subject.__init__(self)
        self._name = name # Set the name of the model
        self._labels = ['Python', 'C++', 'Java', 'Perl', 'Scala', 'Lisp']
        self._data = [10,8,6,4,2,1]

    @property #Getter that gets the labels
    def labels(self):
        return self._labels

    @labels.setter #Setter that sets the labels
    def labels(self, labels):
        self._labels = labels
        self.notify() # Notify the observers whenever somebody changes the labels
        
    @property #Getter that gets the data
    def data(self):
        return self._data

    @data.setter #Setter that sets the labels
    def data(self, data):
        self._data = data
        self.notify() # Notify the observers whenever somebody changes the data
        

# This is the 'standard' view/observer which also acts as an 'abstract' class whereby deriving Bar/Chart/Table views override the update() method
# This 'abstracted' layer is always shown in examples but is important to demonstrate potential polymorphic behaviour of update()
class View():
    
    def __init__(self, name=""):
         self._name = name #Set the name of the Viewer

    def update(self, subject): #Alert method that is invoked when the notify() method in a concrete subject is invoked
        print("Generalised Viewer '{}' has: \nName = {}; \nLabels = {}; \nData = {}".format(self._name, subject._name, subject._labels, subject._data))

# Table 'chart' viewer/ConcreteObserver - overrides the update() method
class TableView(View):
    
    def update(self, subject): #Alert method that is invoked when the notify() method in a concrete subject is invoked
        fig = plt.figure(dpi=80)
        ax = fig.add_subplot(1,1,1)
        table_data = list(map(list,zip(subject._labels, subject._data)))
        table = ax.table(cellText=table_data, loc='center')
        table.set_fontsize(14)
        table.scale(1,4)
        ax.axis('off')
        plt.show()

# Bar chart viewer/ConcreteObserver - overrides the update() method
class BarView(View):
        
    def update(self, subject): #Alert method that is invoked when the notify() method in a concrete subject is invoked
        objects = subject._labels
        y_pos = np.arange(len(objects))
        performance = subject._data
        plt.bar(y_pos, performance, align='center', alpha=0.5)
        plt.xticks(y_pos, objects)
        plt.ylabel('Usage')
        plt.title('Programming language usage')
        plt.show()

# Pie chart viewer/ConcreteObserver - overrides the update() method
class PieView(View):

    def update(self, subject): #Alert method that is invoked when the notify() method in a concrete subject is invoked
        # Pie chart, where the slices will be ordered and plotted counter-clockwise:
        labels = subject._labels
        sizes = subject._data
        explode = (0.1, 0, 0, 0, 0, 0)  # only "explode" the 1st slice
        fig1, ax1 = plt.subplots()
        ax1.pie(sizes, explode=explode, labels=labels, autopct='%1.1f%%', shadow=True, startangle=90)
        ax1.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
        plt.show()

#Let's create our subjects
m1 = Model("Model 1")
m2 = Model("Model 2") # This is never used!

#Let's create our observers
v1 = View("1: standard text viewer")
v2 = TableView("2: table viewer")
v3 = BarView("3: bar chart viewer")
v4 = PieView("4: pie chart viewer")

#Let's attach our observers to the first model
m1.attach(v1)
m1.attach(v2)
m1.attach(v3)
m1.attach(v4)

# Let's just call the notify() method to see all the charts in their unchanged state
m1.notify()

# Now Let's change the properties of our first model
# Change 1 triggers all 4 views and updates their labels
m1.labels = ['C#','PHP','JavaScript','ASP','Python','Smalltalk']
# Change 2 triggers all 4 views and updates their data
m1.data = [1,18,8,60,3,1]
````

# <font color="red">The Prototype Design Pattern - Exercise 16</font> #

## <font color="red">Intent - Exercise 16</font> ##

Create new objects by copying the prototype, which is an existing object, which is an efficient way to create new instances of an existing object, rather than creating it from scratch.

## <font color="red">Motivation - Exercise 16</font> ##

The prototype design pattern is motivated by scenarios where it would be cheaper, require less computing power, and take less time to create a copy of an object instead of creating from scratch. It could be used very effectively in GUI creation for programs, as we could create prototyped elements for things such as buttons, text entry fields and more, which we can then clone and customise it to fit the needs of the GUI, rather than creating all these elements from the ground up, which would take much more computing power and time, both from the program itself running, and the time the developers would take to implement everything, which would also increase the costs of the program.

The client will ask the prototype to clone itself by calling the Clone() method.

## <font color="red">Implementation - Exercise 16</font> ##

When implementing prototypes, there are a few things we should take into consideration. The first is the implementation of the clone operation. Clones can either be deep or shallow clones. A deep clone will clone the variables in the cloning object, whereas a shallow clone will share the variables between itself and the cloning object. A shallow clone is often enough and is simpler to implement, however a deep clone would be better for a more complex object, to keep itself and the original independent.


There are also instances where we would need to initialise the internal states of the clones as these values can’t be passed to the Clone() method, so the program will not function properly if we don’t initialise these values.

When the amount of prototypes in a system isn’t a set number, meaning they can be created and destroyed so the amount will vary, we use a prototype manager. The prototype manager is a registry, that the clients can access. Instead of the clients themselves managing the prototypes, they will store and retrieve them from this registry.


## <font color="red">Sample Code - Exercise 16</font> ##

The following is a basic example of the prototype design pattern, where we create an initial object of a person including their name and age, which is the prototype, and then use copy to create shallow clones of this object and change the names and ages of these new objects.

````python
import copy

#Abstract Prototype Class
class Prototype:
    def clone(self):
        pass

#Concrete Prototype Class
class ConcretePrototype(Prototype):
    def __init__(self, name, age):
        self.name = name
        self.age = age

    def clone(self):
        #Using copy to create a shallow clone of the object
        return copy.copy(self)

    def __str__(self):
        #returns the name and age to be printed
        return f"Name:{self.name}, Age:{self.age}"

#Create an initial object which is the prototype
prototype = ConcretePrototype("Ethan Smith", 21)
    
#Uses the clone() method to create 2 shallow clones of the prototype
clone1 = prototype.clone()
clone2 = prototype.clone()

#Modifying the first clone
clone1.name = "Jane Williams"
clone1.age = 25

#Modifying the second clone
clone2.name = "Josh Brown"
clone2.age = 30

#Print the original prototype and the 2 clones to show the original prototype hasn't changed
print("Original Prototype:", prototype)
print("Clone 1:", clone1)
print("Clone 2:", clone2)
````

## <font color="red">Evaluation - Exercise 16</font> ##

The prototype design pattern has many benefits including the ability to and remove objects at runtime, creating new objects by changing values or structure, and reduced subclassing, as the prototype design pattern doesn’t require a creator class hierarchy.

The main drawback of the prototype design pattern is that every subclass of the prototype has to implement the clone operation, which can be difficult when the classes already exist, or they have objects that don’t support copying or have circular references.

The prototype design pattern is best used when you will have lots of similar objects in a program, that use the same variables and structure, as instead of wasting resources creating an object from scratch we can clone the prototype, which will contain the variables and structure needed for our new object.

# <font color="red">The State Design Pattern - Exercise 17</font> #

## <font color="red">Intent - Exercise 17</font> ##

Allow an object to change its behaviour when its internal state changes, allowing it to change its behaviour at runtime depending on its internal state.

## <font color="red">Motivation - Exercise 17</font> ##

The state design pattern is motivated by scenarios where an object could have multiple states that would require different behaviour depending on that state. For example, an object may have a not started, in progress, or finished state, and when we output something, we want it to change depending on the objects state. This could also be achieved by creating multiple conditional if statements, however the state design pattern allows you to encapsulate each state into a different class, keeping your code more maintainable and easier to read.

## <font color="red">Structure - Exercise 17</font> ##

Diagram from Design patterns CD: Elements of reusable object-oriented software (pg. 315)

![State_Structure.png](State_Structure.png)

- Context: Defines the interface of interest to clients and maintains an instance of the ConcreteState subclass that defines the state that the object is currently in.
- State: Interface that represents all the different states that Context can be in.
- ConcreteState: Implements the specific behaviour that is associated with each state.


## <font color="red">Collaborations - Exercise 17</font> ##

The state specific behaviour is given to ConcreteState by Context and when the state changes, context updates its reference to the new ConcreteState.

## <font color="red">Implementation - Exercise 17</font> ##

When we define what triggers a state transition, this can be done withon Context, if the criteria for the transition is fixed. It would be more flexible to add interfaces to the State subclasses to allow them to determine the state transitions, which would require them being able to set the current state of Context. This decentralising of the state transitions makes it easier to define new State subclasses, however each subclass becomes more dependent on each other as they will need information about each other to be able to complete state transitions.

Alternatively, you could create a table, which defines the criteria for each state transition, as well as what each state transition entails, allowing you to easily change the criteria by just changing table data instead of code, however it would be less efficient to perform a table lookup rather than call a function and it would likely to be harder to understand what the table details compared to the code in a function.

It is also important to consider whether state object will be created when needed and then destroyed after or whether they are created beforehand and then never destroyed. If you don’t know when and what states will be entered at runtime, the first approach is likely better as it will save memory as Context doesn’t have to keep references to each state, however if the same states are entered over and over, the second approach is likely better, as it would be more efficient to keep these references to each state rather than creating each state object every time we enter that state.


## <font color="red">Sample Code - Exercise 17</font> ##

The following is a basic example of the state design pattern, where we have a document, and we move it through three different states, draft, moderation and published, representing how far along the document’s creation and publication is, and we output a different message depending on the current state.

````python
#Concrete States

#The draft state, which is the first state of a document
class DraftState:
    def publish(self, document):
        #Calls set_state and changes to the next state when called, which is moderation
        print("Moving from Draft to Moderation.")
        document.set_state(ModerationState())

    def __str__(self):
        #returns draft to be printed
        return "Draft"

#The moderation state, which is the second state of a document
class ModerationState:
    def publish(self, document):
        #Calls set_state and changes to the next state when called, which is published
        print("Moving from Moderation to Published.")
        document.set_state(PublishedState())

    def __str__(self):
        #returns moderation to be printed
        return "Moderation"

#The published state, which is the final state of a document
class PublishedState:
    def publish(self, document):
        #Published is the final state, so it cannot be changed further so set_state isn't called
        print("Document is already Published. No further changes.")

    def __str__(self):
        #returns moderation to be printed
        return "Published"

#Context
class Document:
    def __init__(self):
        #Sets the current state to draft, as this is the starting state of each document
        self.state = DraftState()
        print(f"Document created in state: {self.state}")

    def set_state(self, new_state):
        #chnages the state to the next one when called
        self.state = new_state
        print(f"State changed to: {self.state}")

    def publish(self):
        #Calls the publish method from the current state
        print(f"Current state: {self.state}")
        self.state.publish(self)

doc = Document()
doc.publish()  # Moves to moderation
doc.publish()  # Moves to published
doc.publish()  # Already published
````

## <font color="red">Evaluation - Exercise 17</font> ##

The state design pattern allows for a lot of flexibility, as it is very easy to add new states, as you can simply create a new class for a new state, rather than having to add to your existing code. This also provides more clarity as it makes the code much more readable as it is easier to determine what is happening in each state, as they are broken up into their own classes. This also makes the code simpler, as all the state-specific behaviour is inside the classes.

The state design pattern is best used when you have an object that can multiple different states, that requires different behaviour depending on these states. It can be much more efficient to use the state design pattern compared to creating conditional statements, and its efficiency in comparison only increases as the number of states increases.
